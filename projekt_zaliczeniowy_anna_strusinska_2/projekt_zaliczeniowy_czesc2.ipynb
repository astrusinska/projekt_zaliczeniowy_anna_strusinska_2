```python
# 4.	Wyświetl tweety użytkownika, który założył konto najwcześniej (spośród wszystkich użytkowników w datasecie) (kolumna user_created_at).

df_twitter["user_created_at"]
```

```python
df_twitter["user_created_at"] = pd.to_datetime(df_twitter["user_created_at"])
```

```python
first_time = df_twitter["user_created_at"].min()
```

```python
first_user = df_twitter[df_twitter["user_created_at"] == first_time]
```

```python
print(first_user["text"])
```

```python
# 5.	Wyświetl tweety użytkownika. który ma najwięcej followersów.

most_followers = df_twitter["user_followers_count"].max()
```

```python
most_popular_user = df_twitter[df_twitter["user_followers_count"] == most_followers]
```

```python
print(most_popular_user["text"])
```

```python
# 7.	Podaj, w który dzień tygodnia najczęściej były publikowane tweety z datasetu

df_twitter["created_at"] = pd.to_datetime(df_twitter["created_at"])
```

```python
df_twitter["weekday"] = df_twitter["created_at"].dt.day_name()
```

```python
 most_frequent_weekday = df_twitter["weekday"].mode()
 print(most_frequent_weekday)
```
