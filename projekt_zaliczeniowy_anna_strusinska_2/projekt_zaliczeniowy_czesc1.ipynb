```python
# 6.	UsuÅ„ wszystkie sÅ‚owa oznaczone jako stopwords i przekaÅ¼ oczyszczony tekst do nowej kolumny o nazwie text_without_stopwords

!python -m spacy download pl_core_news_lg
```

```python
import spacy
```

```python
nlp = spacy.load("pl_core_news_lg")
```

```python
def delete_stopwords(text):
  doc = nlp(text)
  wynik = " ".join([token.text for token in doc if not token.is_stop])
  return wynik
```

```python
df_twitter["text_without_stopwords"] = df_twitter["text"].apply(delete_stopwords)
```
